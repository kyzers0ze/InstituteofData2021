```python
## Import Libraries
import regex as re
import numpy as np
from urllib.parse import unquote
import urllib3
from bs4 import BeautifulSoup
import time
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
categories = ['worldnews','businessnews','technologynews']
page_nums = list(range(1,31))
urls = []

for category in categories:
    for page_num in page_nums:
        url = 'https://www.reuters.com/news/archive/{category}?view=page&page={page_num}&pageSize=10'.format(category=category,
                                                                                                             page_num = page_num)
        urls.append(url)
urls
```

```python
browser = webdriver.Chrome('C:\chromedriver_win32\chromedriver.exe')
for url in urls:
    data = []
    
    browser.get(url)
    browser.execute_script(
        "window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;"
    )
    time.sleep(5)

    results = browser.find_elements_by_xpath(
        "//*[@class='story-content']"
    )
    print('Number of results', len(results))

    for result in results:
        print(result)
        product_name = result.text
        link = result.find_element_by_tag_name('a').get_attribute("href")
        data.append({"product" : product_name, "link" : link})
browser.quit()
```
