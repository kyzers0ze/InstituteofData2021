```python
## Import Libraries
import regex as re
import numpy as np
from urllib.parse import unquote
import urllib3
from bs4 import BeautifulSoup
import time
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
categories = ['worldnews','businessnews','technologynews']
page_nums = list(range(1,31))
urls = []

for category in categories:
    for page_num in page_nums:
        url = 'https://www.reuters.com/news/archive/{category}?view=page&page={page_num}&pageSize=10'.format(category=category,
                                                                                                             page_num = page_num)
        urls.append(url)
urls
```

```python
browser = webdriver.Chrome('C:\chromedriver_win32\chromedriver.exe')
article_count = 0
data = []
for url in urls:
  
    browser.get(url)
    browser.execute_script(
        "window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;"
    )
    time.sleep(5)

    results = browser.find_elements_by_xpath(
        "//*[@class='story-content']"
    )
    print('Number of results', len(results))
    article_count += len(results)
    print('Article count: ', article_count)
    for result in results:
        print(result)
        product_name = result.text
        link = result.find_element_by_tag_name('a').get_attribute("href")
        data.append({"product" : product_name, "link" : link})
browser.quit()
```

```python
df = pd.DataFrame(data)
df
```

```python
article_content = []
count = 0
for url in df.link:

    link = url

    http = urllib3.PoolManager()
    r = http.request('GET', link)
    if r.status == 200:
        page = r.data
        print('Type of the variable \'page\':', page.__class__.__name__)
        print('Page Retrieved. Request Status: %d, Page Size: %d' % (r.status, len(page)))
    else:
        print('Some problem occurred. Request Status: %s' % r.status)
    soup = BeautifulSoup(page, 'html.parser')

    para_class = 'Paragraph-paragraph-2Bgue ArticleBody-para-TD_9x'

    text = ''
    article = soup.find_all("p",{"class":para_class})

    for tag in article:
        text += tag.getText()
    count += 1
    print("Article number...{}".format(count))
    print('Appending to text...')
        
    article_content.append(text)
```

```python
article_content[0]
```

```python
article_df = pd.DataFrame({'article_content':article_content})
article_df
```

```python
combined_df = pd.concat([df,article_df],axis=1)
combined_df
```

```python
combined_df.to_csv('reuters_articles.csv')
```
