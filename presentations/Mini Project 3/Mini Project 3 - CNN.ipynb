```python
## Import Libraries
import regex as re

from urllib.parse import unquote
import urllib3
from bs4 import BeautifulSoup
import time
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
initial_query= 50
current_article_number = 0
current_page = 1

desired_page = 20

urls = []

for i in range(desired_page):
    url = "https://edition.cnn.com/search?size={initial_query}&q=*&sort=newest&from={current_article_number}&page={current_page}&type=article".format(
    initial_query=initial_query,
    current_article_number=current_article_number,
    current_page = current_page)
    current_article_number += 50
    current_page += 1
    urls.append(url)
urls
```

```python
data = []
for url in urls:
    browser = webdriver.Chrome('C:\chromedriver_win32\chromedriver.exe')
    base_url = url
    browser.get(base_url)
    browser.execute_script(
        "window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;"
    )
    time.sleep(5)

    results = browser.find_elements_by_xpath(
        "//*[@class='cnn-search__result-headline']"
    )
    print('Number of results', len(results))
    # loop over results
    for result in results:
        product_name = result.text
        link = result.find_element_by_tag_name('a')
        product_link = link.get_attribute("href")
        # append dict to array
        data.append({"product" : product_name, "link" : product_link})
    browser.quit()
```

```python
cnn_links = [x['link'] for x in data]
cnn_titles = [ x['product'] for x in data]
cnn_links, cnn_titles
```
