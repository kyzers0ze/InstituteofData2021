```python
import pandas as pd
import numpy as np
```

```python
csv_files = ['breitbart_articles_titles.csv','cnn_articles.csv','daily_wire.csv','reuters_articles.csv']
```

```python
data_frames = []
for file in csv_files:
    data_frames.append(pd.read_csv(file))
```

```python
df = pd.concat(data_frames)
df = df.drop('Unnamed: 0',axis=1)
df
```

```python
df = df.dropna()
df
```

```python
df = df.rename(columns={'product':'title'})
df
```

```python
df.source.value_counts()
```

```python
df = df.drop_duplicates()
df.source.value_counts()
```
