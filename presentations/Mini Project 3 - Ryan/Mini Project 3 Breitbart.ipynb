```python
def replace_us_space(text):
    us_space_pattern = ' ?(u s )'
    us_matcher = re.search(us_space_pattern, text)
    if us_matcher:
        new_text = text.replace("u s", "us")
        return new_text
    else:
        return text 
    
def is_article_link(category, text):
    
    return category in str(text) and str(text) != '/' + category + '/' and 'page' not in str(text)
    

def get_cleaned_links(url,category):    
    http = urllib3.PoolManager()
    r = http.request('GET', url)
    if r.status == 200:
        page = r.data
        print('Type of the variable \'page\':', page.__class__.__name__)
        print('Page Retrieved. Request Status: %d, Page Size: %d' % (r.status, len(page)))
    else:
        print('Some problem occurred. Request Status: %s' % r.status)
    soup = BeautifulSoup(page, 'html.parser')
    article = soup.find_all("a")
    tag_list = [t.get('href') for t in article]
    cleaned_links = [text for text in tag_list if is_article_link(category, text)]
    return list(set(cleaned_links))    
```

```python
results = []
for url in urls:
    if 'politics' in url:
        result = get_cleaned_links(url, 'politics')
        results.append(result)
    elif 'economy' in url:
        result = get_cleaned_links(url, 'economy')
        results.append(result)
    elif 'asia' in url:
        result = get_cleaned_links(url, 'asia')
        results.append(result)
    elif 'tech' in url:
        result = get_cleaned_links(url, 'tech')
        results.append(result)
    else:
        result = get_cleaned_links(url,'europe')
        results.append(result)
results
```

```python
breitbart_links = []
for result in results:
    for i in range(len(result)):
        breitbart_links.append(result[i])
```

```python
article_content = []
titles = []
for breitbart_link in breitbart_links:
    if 'https://www.breitbart.com' in breitbart_link:
        link = breitbart_link
    else:
        link = 'https://www.breitbart.com' + breitbart_link
    print(link)
    http = urllib3.PoolManager()
    r = http.request('GET', link)
    if r.status == 200:
        page = r.data
        print('Type of the variable \'page\':', page.__class__.__name__)
        print('Page Retrieved. Request Status: %d, Page Size: %d' % (r.status, len(page)))
    else:
        print('Some problem occurred. Request Status: %s' % r.status)
    soup = BeautifulSoup(page, 'html.parser')

    text = ''
    article = soup.find_all("p")
    for tag in article:
        try:
            if tag['class'] == 'rmoreabt':
                continue
            else:
                text = text + tag.getText()
        except:
            text = text + tag.getText()
    article_content.append(text)
    title = soup.find('h1').getText()
    titles.append(title)
    print('appending article content: ',len(article_content))
    print('appending title: ', len(titles))
```

```python
df = pd.DataFrame({'link':breitbart_links,'title':titles,'article_content':article_content})
df.tail(10)
```
