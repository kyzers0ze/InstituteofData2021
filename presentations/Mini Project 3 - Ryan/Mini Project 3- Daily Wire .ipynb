```python
## Import Libraries
import regex as re
import numpy as np
from urllib.parse import unquote
import urllib3
from bs4 import BeautifulSoup
import time
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
page_nums = list(range(1,50))
urls = []

for page_num in page_nums:
    url = 'https://www.dailywire.com/search/news?query=%2A&page={page_num}'.format(page_num = page_num)
    urls.append(url)
urls
```

```python
http = urllib3.PoolManager()

r = http.request('GET', urls[1])
if r.status == 200:
    page = r.data
    print('Type of the variable \'page\':', page.__class__.__name__)
    print('Page Retrieved. Request Status: %d, Page Size: %d' % (r.status, len(page)))
else:
    print('Some problem occurred. Request Status: %s' % r.status)
soup = BeautifulSoup(page, 'html.parser')
```

```python
data = []
browser = webdriver.Chrome('C:\chromedriver_win32\chromedriver.exe')
for url in urls:  
    browser.get(url)
    browser.execute_script(
        "window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;"
    )
    time.sleep(5)

    results = browser.find_elements_by_xpath(
        "//*[@class='css-k008qs']"
    )
    print('Number of results', len(results))

    for result in results:
        print(result)
        product_name = result.text
        link = result.get_attribute("href")
        data.append({"product" : product_name, "link" : link})
browser.quit()
```

```python
df = pd.DataFrame(data)
df.link.value_counts()
```

```python
article_content = []
count = 0
unwanted_count = 0
for url in df.link:
    
    r = http.request('GET', url)
    if r.status == 200:
        page = r.data
        print('Type of the variable \'page\':', page.__class__.__name__)
        print('Page Retrieved. Request Status: %d, Page Size: %d' % (r.status, len(page)))
    else:
        print('Some problem occurred. Request Status: %s' % r.status)
    soup = BeautifulSoup(page, 'html.parser')
    article = soup.find_all('p')
    text = ''
    unwanted_text = 'The Daily Wire is one of Americas fastest-growing conservative media companies'
    
    for tag in article:   
        if unwanted_text not in tag.getText():
            text += tag.getText()
            print('Adding to article content...')
        else:
            unwanted_count += 1
            print('Encoutered unwanted text')
    count += 1
    
    print(count,unwanted_count,url)
    
    article_content.append(text)
```

```python
df_content = pd.DataFrame({'article_content':article_content})
df_content
```

```python
df = pd.concat([df,df_content],axis=1)
df
```

```python
df.to_csv('daily_wire.csv')
```
