# ROC Curve Testing - Ryan

The purpose of this notebook is to see if we get a poor performing model, can we do the opposite of what the model predicts and obtain a better performing model.

The metric used to determine this would be the Receiver Operating Characteristic or ROC for the regular model and the inverse model.

Inverse model meaning that if the regular model predicts 1, the inverse model would predict 0.

The hypothesis is that the the ROC curve would be flipped along the y=x axis.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
def get_FPR(FP,TN):
    return FP / float(FP + TN)
def get_TPR(TP,FN):
    return TP/ float(TP + FN)
def get_FPR_TPR_arr(threshold_step=0.01,flip_prediction=False):
    fpr_arr, tpr_arr = [], []
    threshold_arr = np.arange(0,1,threshold_step)
    for threshold in threshold_arr:
        pred_list = [1 if x >= threshold else 0 for x in model_pred]
        if flip_prediction:
            pred_list = [0 if x == 1 else 1 for x in pred_list]
        pred_actual_comparison = zip(pred_list,actual_pred)
        TP,FP,FN,TN = 0,0,0,0
        for pred,actual in pred_actual_comparison:
            if actual == 1 and pred == 1:
                TP += 1
            elif actual == 0 and pred == 1:
                FP += 1
            elif actual == 1 and pred == 0:
                FN += 1
            else:
                TN += 1
        fpr_arr.append(get_FPR(FP,TN))
        tpr_arr.append(get_TPR(TP,FN))
    return fpr_arr,tpr_arr

def get_auc(fpr_arr,tpr_arr):
    width_of_rectangle = 1 / len(fpr_arr)
    sum_rectangle = 0
    for i in range(len(fpr_arr)):
        sum_rectangle += tpr[i]*width_of_rectangle
    return sum_rectangle      
```
