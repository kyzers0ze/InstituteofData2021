Format for Pokemon API request is `https://pokeapi.co/api/v2/pokemon/{id}or{pokemon_name}`


```python
#First 3 pokemon
pokemon_first_3_id = [1,2,3]
url = 'https://pokeapi.co/api/v2/pokemon/'
pokemon_json_list = []
for pokemon_id in pokemon_first_3_id:
    pokemon_json_list.append(requests.get(url+str(pokemon_id)))
pokemon_json_list
```

```python
for json_list in pokemon_json_list:
    print(json.loads(json_list.text).keys())
```

```python
pokemon_dict = {}
i=0
for json_list in pokemon_json_list:
    print(json.loads(json_list.text)['name'])
    pokemon_dict[i] = json.loads(json_list.text)['name']
    i = i+1
```

```python
pokemon_dict
```

```python
ability_list = []
for json_list in pokemon_json_list:
    ability_list.append(json.loads(json_list.text)['abilities'])
ability_list
```

```python
for i in range(len(ability_list)):
    print(pokemon_dict[i],ability_list[i])
```
