# Statistics - Part 1

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline
```

```python
help(plt.pie)
```

```python
# plot frequency of some observations using pie chart
fig, ax = plt.subplots()
labels = 'Developers', 'Data Engineers', 'Data Scientists', 'Business Analysts'
frequency = [15, 32, 12, 7]

# only "explode" the 3rd slice (i.e. 'Data Scientist')
# show labels and the value of the wedge as a % to 1 decimal place)
explode = (0, 0, 0.2, 0) 

plt.pie(frequency,explode=explode,labels=labels,autopct='%1.1f%%')
plt.title("Number of Data Related Jobs")
plt.legend()
# Use 'equal' axis for aspect ratio to ensure that pie is drawn as a circle
ax.axis('equal')
```

```python
help(plt.bar)
```

```python
# plot frequency of some observations using bar chart

objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
performance = [10,8,6,4,2,1]
 
# Create a bar plot with title 'Programming language usage' & y-axis label 'Usage'
color = ['red','blue','green','purple','red','orange']
plt.bar(objects,performance,color=color)
plt.title('Programming language usage',fontsize=20)
plt.ylabel("Usage").set_rotation(0)
```
