# 10. [BONUS] Create a probplot against normal distribution

Check following methods:

1. stats.loggamma.rvs
2. stats.probplot
3. stats.norm
4. stats.boxcox

````python
from scipy import stats
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2, 2)

# log gamma 
x_loggamma = stats.loggamma.rvs(5, size=500) + 5
axs[0, 0].plot(x_loggamma)
axs[0, 0].set_title('Log Gamma')
# boxcox
x_boxcox, _ = stats.boxcox(x_loggamma)

axs[0, 1].plot(x_boxcox)
axs[0, 1].set_title('Boxcox')
# norm

stats.probplot(x_loggamma, dist=stats.loggamma, sparams=(2.5,), plot=axs[1, 1])
plt.show()
````

# 11. [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem
2. scipy.stats.t.ppf

````python
import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    # Calculate m (mean), se (std. error of mean), h (distance from mean to C.I.)
    m, se = np.mean(data), scipy.stats.sem(data)
    h = scipy.stats.t.ppf(confidence,data)
    
    return m, m-h, m+h

data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
print(mean_confidence_interval(data))

np.mean(data)
````
