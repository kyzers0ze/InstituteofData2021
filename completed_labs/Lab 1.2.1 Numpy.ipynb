# Numpy Exploration

````python
import numpy as np
import matplotlib.pyplot as plt

# get matplot to show charts inline
%matplotlib inline
````

````python
height= np.round(np.random.normal(1.75, 0.20, 7000), 2)
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight))
````

````python
print('height:\n',height)
print('weight:\n',weight)
print('population:\n',population)
````

````python
# **1. Find Dimension of Array**

# Find dimention of each array: height, weight, population
# ```
# Expected Ouput
# np.ndim(height): 1
# np.ndim(weight): 1
# np.ndim(population): 2
# ```

print(np.ndim(height))
print(np.ndim(weight))
print(np.ndim(population))
````

````python
# **2. Find Shape of Array**

# Find shape of each array: height, weight, population
# ```
# Expected Ouput
# np.shape(height) : (7000,)
# np.shape(weight) : (7000,)
# np.shape(population) : (7000, 2)
# ```

print(np.shape(height))
print(np.shape(weight))
print(np.shape(population))
````

````python
#**3. Find Size of Array**

#Find shape of each array: height, weight, population
#```
# Expected Ouput:
#np.size(height) : 7000
#np.size(weight) : 7000
#np.size(weight) : 14000
#```

print(np.size(height),np.size(weight),np.size(population))
````

````python
#**4. Find Size of Items of Each Array**

#Find size of items each array: height, weight, population
#```
# Expected Output:
#height.itemsize : 8
#weight.itemsize : 8
#population.itemsize : 8
#```
print(height.itemsize,weight.itemsize,population.itemsize)
````

````python
#**4. Find Minimum Value of Each Array**

#Find minimum value of each array: height, weight, population
#```
# Expected Output:
#height.min : 1.0
#weight.min : 3.24
#population.min : 1.0
#```

print(height.min(),weight.min(),population.min())
````

````python
#**5. Find Maximum Value of Each Array**

#Find maximum value of each array: height, weight, population
#```
# Expected Output
#height.max() : 2.51
#weight.max() : 132.17
#population.max() : 132.17
#```

print(height.max(),weight.max(),population.max())
````

````python
#**6. Find Mean Value of Each Array**

#Find maximum value of each array: height, weight, population
#```
# Expected Output
#height.mean() : 1.7494871428571428
#weight.mean() : 60.41684428571429
#population.mean() : 31.083165714285716
#```
print(height.mean(),weight.mean(),population.mean())
````

````python
#**6. Find Standard Deviation of Each Array**

#Find standard deviation value of each array: height, weight, population
#```
# Expected Output
#height.std() : 0.1981923748970856
#weight.std() : 14.821310711612284
#population.std() : 31.090684288943482
#```

print(height.std(),weight.std(),population.std())
````

````python
#**7. Find Total value of Each Array**

#Find total value of each array: height, weight, population
#```
# Expected Output
#height.sum() : 12219.789999999999
#weight.sum() : 422009.95
#population.sum() : 434229.74
#```

print(height.sum(),weight.sum(),population.sum())
````

**8. Create a histogram of height and weight**

````python
def set_shared_ylabel(a, ylabel, labelpad = 0.01):
    """Set a y label shared by multiple axes
    Parameters
    ----------
    a: list of axes
    ylabel: string
    labelpad: float
        Sets the padding between ticklabels and axis label"""

    f = a[0].get_figure()
    f.canvas.draw() #sets f.canvas.renderer needed below

    # get the center position for all plots
    top = a[0].get_position().y1
    bottom = a[-1].get_position().y0

    # get the coordinates of the left side of the tick labels 
    x0 = 1
    for at in a:
        at.set_ylabel('') # just to make sure we don't and up with multiple labels
        bboxes, _ = at.yaxis.get_ticklabel_extents(f.canvas.renderer)
        bboxes = bboxes.inverse_transformed(f.transFigure)
        xt = bboxes.x0
        if xt < x0:
            x0 = xt
    tick_label_left = x0

    # set position of label
    a[-1].set_ylabel(ylabel)
    a[-1].yaxis.set_label_coords(tick_label_left - labelpad,(bottom + top)/2, transform=f.transFigure)


fig, axs = plt.subplots(2)
fig.tight_layout()
axs[0].hist(height,bins=20,color='red')
axs[1].hist(weight,bins=20,color='cyan')

axs[0].title.set_text('Histogram of Height Distribution')
axs[1].title.set_text('Histogram of Weight Distribution')

set_shared_ylabel(axs, 'Number of People')

axs[0].set_xlabel('Height (m)')
axs[1].set_xlabel('Weight (kg)')

axs[0].grid()
axs[1].grid()

fig.subplots_adjust(top=1.5)
````

**9. Create a boxplot of height and weight**

````python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.tight_layout()

ax1.boxplot(height)
ax2.boxplot(weight)

ax1.title.set_text('Height')
ax2.title.set_text('Weight')

ax1.set_ylabel('Height (m)')
ax2.set_ylabel('Weight (kg)')
fig.subplots_adjust(right=1.2)

ax1.grid()
ax2.grid()
````

**10. Create scatterplot of height and weight**

````python
plt.scatter(weight,height,alpha=0.2,color='purple',marker='x')
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("Plot of Height against Weight")
plt.grid()
````
